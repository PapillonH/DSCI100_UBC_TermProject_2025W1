### Data Direct and Potential Issues

#### Observable Issues
- **Missing values:**  
  - There're missing data marked as `NA`.  
  - There're missing raw timestamps (`original_start_time` or `original_end_time`).
- **Outliers:**  
  - `duration_time` is the mutate variable from start_time and end_time, which extremely right-skewed; a few players have over 4 hours while most have < 1 hour.  
- **Formatting inconsistencies:**  
  - `start_time` and `end_time` are stored as character, requiring conversion before numerical operations.  

#### Potential Hidden Issues
- **Sampling bias:**  
  It may present the volunteer bias because players volunteered to join the PLAICraft research server, so the sample likely over-represents active of Minecraft-enthusiast users.
- **Time-zone uncertainty:**  
  The logs may combine users from different regions; if timestamps were stored in local time without time-zone normalization, hourly activity peaks could appear blurred.
- **Device or network:**  
  Multiple accounts on the same machine or reconnections after disconnections could fragment a single play period into multiple short sessions.

```r
players_clean <- players |>
  select(Age, gender, experience, played_hours, subscribe)

players_clean <- players_clean |>
  rename(age = Age) |>
  mutate(
    gender = as.factor(gender),
    experience = as.factor(experience),
    subscribe = as.factor(subscribe)
  )

summary(players_clean)
# glimpse(players_filtered)
```

```r
hist_plot <- players_clean|>
    ggplot(aes(x = played_hours)) +
    geom_histogram(bins = 30, fill = "skyblue", color = "white") +
    labs(title = "Distribution of Total Played Hours",
        x = "Played Hours (hours)",
        y = "Number of Players") +
    theme(text = element_text(size = 20))
hist_plot
```

From the histogram of dirstribution of total play hours, we can tell: most players have played only a few hours, there's only a small group has spent over 100 hours on the server. It's a Strong right skewed distribution.   
The overall pattern also suggests that **playtime could be a strong indicator of engagement**, which is potentially related to whether a player subscribes to the newsletter.

```r
bar_plot <- players_clean|>
    ggplot(aes(x = experience, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(title = "Subscription Rate by Experience Level",
       x = "Experience Level",
       y = "Proportion of Players") +
    scale_y_continuous(labels = scales::percent) +
    theme(text = element_text(size = 20))
bar_plot
```

From the bar plot of subsrcibe status and experience level, we can tell: The majority of players across all experience levels are subscribers. Higher experience levels (Veteran, Pro) show a less poportion of subscribers than lower levels (Amateur, Regular).   
This suggests that **experience level has some predictive power**. In modeling, this variable could still contribute useful information to help predict subscription probability.
